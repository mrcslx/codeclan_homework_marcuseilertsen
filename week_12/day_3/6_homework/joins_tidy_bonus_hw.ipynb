### Question 2

### Question 3

Join the data containing fraud details with transaction details so all rows of both tables are kept. What does the resulting row number tell you?

```python
fraud_base.merge(transaction_base, how='outer', on='transaction_id')
```

It seems that every row of `fraud_base` had a match in `transaction_base`, as both `transaction_base` and the joined dataset have 10000 rows.
